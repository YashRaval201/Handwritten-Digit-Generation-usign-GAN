```python
import random
```

```python
def is_valid(board, num, position):
    # Check row
    for i in range(len(board[0])):
        if board[position[0]][i] == num and position[1] != i:
            return False

    # Check column
    for i in range(len(board)):
        if board[i][position[1]] == num and position[0] != i:
            return False

    # Check box
    box_x = position[1] // 3
    box_y = position[0] // 3

    for i in range(box_y*3, box_y*3 + 3):
        for j in range(box_x*3, box_x*3 + 3):
            if board[i][j] == num and (i,j) != position:
                return False

    return True
```

```python
def fill_board(board):
    find = find_empty(board)
    if not find:
        return True  # If no empty space is found, board is complete
    else:
        row, col = find

    numbers = list(range(1, 10))
    random.shuffle(numbers)  # Randomize the numbers to introduce variety
    for num in numbers:
        if is_valid(board, num, (row, col)):
            board[row][col] = num

            if fill_board(board):
                return True

            board[row][col] = 0

    return False

def find_empty(board):
    for i in range(len(board)):
        for j in range(len(board[0])):
            if board[i][j] == 0:  # 0 means the cell is empty
                return (i, j)
    return None
```

```python
def remove_numbers(board, attempts=40):
    while attempts > 0:
        row = random.randint(0, 8)
        col = random.randint(0, 8)
        while board[row][col] == 0:
            row = random.randint(0, 8)
            col = random.randint(0, 8)
        backup = board[row][col]
        board[row][col] = 0
        board_copy = [r[:] for r in board]
        attempts -= 1

        if not valid_solution(board_copy):
            board[row][col] = backup
            attempts -= 1

def valid_solution(board):
    """ Check if a board is solvable. If not, revert the change. """
    board_copy = [row[:] for row in board]
    try:
        fill_board(board_copy)
        return True
    except:
        return False
```

```python
def print_board(board):
    for i in range(len(board)):
        if i % 3 == 0 and i != 0:
            print("- - - - - - - - - - - -")

        for j in range(len(board[0])):
            if j % 3 == 0 and j != 0:
                print(" | ", end="")

            if j == 8:
                print(board[i][j])
            else:
                print(str(board[i][j]) + " ", end="")
```

```python
def place_number(board, num, position):
    if is_valid(board, num, position):
        board[position[0]][position[1]] = num
        return True
    return False
```

```python
def main():
    board = [[0 for _ in range(9)] for _ in range(9)]
    fill_board(board)
    remove_numbers(board, attempts=40)  # Adjust the 'attempts' for difficulty
    solved = False

    while not solved:
        print_board(board)
        try:
            row = int(input("Enter row (1-9): ")) - 1
            col = int(input("Enter column (1-9): ")) - 1
            num = int(input("Enter number (1-9): "))

            if place_number(board, num, (row, col)):
                print("Number placed!\n")
            else:
                print("Invalid move, try again.\n")
        except ValueError:
            print("Please enter a valid number.\n")

        if all(all(cell != 0 for cell in row) for row in board):
            solved = True

if __name__ == "__main__":
    main()
```
